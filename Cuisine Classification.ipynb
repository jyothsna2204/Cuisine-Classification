```python
data.drop(['Restaurant ID', 'Restaurant Name', 'City','Address', 'Locality', 'Locality Verbose','Currency'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Cuisines'].fillna('Unknown', inplace=True)
```

```python
data['Cuisine_f'] = data['Cuisines'].str.split(', ').str[0]
data.head()
```

```python
data.shape
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encoder=LabelEncoder()
for i in ["Rating color","Rating text"]:
    data[i]=label_encoder.fit_transform(data[i])
```

```python
b=["Has Table booking","Has Online delivery","Is delivering now","Switch to order menu"]
for i in b:
    data[i]=data[i].map({'Yes':1,'No':0})
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(tokenizer=lambda x: x.split(', '))
cuisine_matrix = vectorizer.fit_transform(data['Cuisines']).toarray()
cuisine_df = pd.DataFrame(cuisine_matrix, columns=vectorizer.get_feature_names_out())
```

```python
x=pd.concat([data.drop(columns=["Cuisines','Cuisine_f"]),cuisine_df],axis=1)
y=data["Cuisine_f"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report

y_pred=model.predict(x_test)
print("CLASSIFICATION REPORT:")
print(classification_report(y_test, y_pred))
```
